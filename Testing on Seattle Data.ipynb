```python
xmls = []
for i in range(5):
    filename = 'data/seattle_' + str(i + 1) + '.xml'
    with open(filename) as file:
        xmls.append(file.read())
```

```python
from schemamatching import compare_xmls
```

```python
%matplotlib inline
import seaborn as sns
```

```python
def compare(i, j):
    results = compare_xmls(xmls[i], xmls[j])
    return sns.heatmap(results, annot=True)
```

```python
compare(0, 1)
```

```python
compare(3, 4)
```

```python
compare(0, 4)
```
