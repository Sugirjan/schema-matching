```python
from xml.etree import ElementTree
def write_countries_xml_to_file(countries_df, filename):
    root = ElementTree.Element('Countries')
    for i in range(len(countries_df)):
        country = ElementTree.SubElement(root, 'Country')
        for field in ['name', 'languages', 'area', 'capital', 'currency', 'callingCode', 'latlng']:
            sub_element = ElementTree.SubElement(country, field)
            sub_element.text = str(countries_df.loc[i, field])
    root_element = ElementTree.ElementTree(root)
    root_element.write(filename)
```

```python
write_countries_xml_to_file(countries, 'countries-full.xml')
```
